```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("/content/traffic_accidents.csv")
```

```python
df.describe
```

```python
df
```

```python
df.isnull().sum()
```

```python
df['crash_date']=pd.to_datetime(df["crash_date"], format='%m/%d/%Y %I:%M:%S %p')
df["Hours"]=df["crash_date"].dt.hour
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x='Hours',data=df)
plt.title("Accidents by Hours Of Day")
plt.xlabel("Hour of Day")
plt.ylabel("Number of Accidents")
plt.tight_layout()
plt.show()
```

```python
print (df["weather_condition"].value_counts().head(10))
```

```python
df.isnull().sum()
```

```python
df = df[df['weather_condition'].notna()]
```

```python
top_weather=df["weather_condition"].value_counts().head(10)
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x=top_weather.values,y=top_weather.index,color="red")
plt.title("Top 10 Weather Conditions")
plt.xlabel("Numbers of Accidents")
plt.ylabel("Weather Conditions")
plt.show()
```

```python
df['roadway_surface_cond'].value_counts()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x="roadway_surface_cond",data=df)
plt.title("Road Surface Conditions")
plt.xlabel("Road Surface Conditions")
plt.ylabel("Number of Accidents")
plt.show()
```

```python
numeric_columns = df.select_dtypes(include=['int64', 'float64']).columns
correlation_matrix = df[numeric_columns].corr()

plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap of Numerical Variables')
plt.show()
```
